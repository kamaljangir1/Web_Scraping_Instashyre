```python
#Import Important Python Libraries
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
import time
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#Create a Function for Driver, in this We Call the Chrome Browser and Maximize the window
def Setup_Driver():
    Driver = webdriver.Chrome()
    Driver.maximize_window()
    return Driver
```

```python
#Scrap the Data and Store in Variables List
#----------------------------------------------------

Driver = Setup_Driver() #Call the Setup_Driver Function, that we Created Before
Driver.get("https://www.instahyre.com/python-jobs") #Store Website URL in Driver Variable, We Extract the Data from this URL

time.sleep(5) #Sleep is important to Optimize the System

    #Create Variables List, Where We Append our Data that We will Extract from WebSite
Name_List = []
Location_List = []
Found_List = []
Employees_List = []
About_List = []
Skills_List = []
Links_List = []

    #Run a Loop so that We Extract the Data of Multiple Pages.
for i in range(1,51):
    #By Using Xpath, We Store Particular path in Particular Variable like "Name"
    Name = Driver.find_elements("xpath","//div[@class='employer-details col-md-10 col-sm-8 col-xs-10']/a/div/div")
    Location = Driver.find_elements("xpath","//div[@class='employer-details col-md-10 col-sm-8 col-xs-10']/a//div[@class='employer-locations']/span[@class='info ng-scope']")
    Found = Driver.find_elements("xpath","//div[@class='employer-details col-md-10 col-sm-8 col-xs-10']//div[@class='employer-info']/span[1]")
    Employees = Driver.find_elements("xpath","//div[@class='employer-details col-md-10 col-sm-8 col-xs-10']//div[@class='employer-info']//span[@class='ng-scope'][2]")
    About = Driver.find_elements("xpath","//div[@class='employer-details col-md-10 col-sm-8 col-xs-10']//div[@class='employer-notes ng-binding ng-scope']")
    Skills = Driver.find_elements("xpath","//div[@class='employer-details col-md-10 col-sm-8 col-xs-10']//div[@class='job-skills ng-scope']")
    Links = Driver.find_elements("xpath","//div[@class='employer-details-mobile col-xs-10']/a")

    #Run a Loop, so that one by one our text Data Append on Variables that we Created Before
    #We Use Try and Except Function, So if There is nothing to Append, it append "Error" String. At The End our Lenght of List Will match
    for j in range(len(Name)):
        try:
            Name_ = Name[j].text
            Name_List.append(Name_)
        except Exception:
            Name_List.append("Error")
          
        try:
            Location_ = Location[j].text
            Location_List.append(Location_)
        except Exception:
            Location_List.append("Error")

        try:
            Founds_ = Found[j].text
            Found_List.append(Founds_)
        except Exception:
            Found_List.append("Error")
        
        try:
            Employee_ = Employees[j].text
            Employees_List.append(Employee_)
        except Exception:
            Employees_List.append("Error")

        try:
            About_ = About[j].text
            About_List.append(About_)
        except Exception:
            About_List.append("Error")

        try:
            Skills_ = Skills[j].text
            Skills_List.append(Skills_)
        except Exception:
            Skills_List.append("Error")

        try:
            Links_ = Links[j].get_attribute("href")
            Links_List.append(Links_)
        except Exception:
            Links_List.append("Error")

    #By using Click Function, We Move From one page to Another page
    element = Driver.find_element("xpath","//div[@class='pagination ng-scope']//li[12]") #Path of NEXT Button
    element.click()

    time.sleep(5)

Driver.close() #When all the Data is Extacted, we Close the Chrome Window.
   
```

```python
#Create the DataFrame Using Pandas and Store all the Data in DataFrame
Data = []
Data = pd.DataFrame(Data)
Data["Name"] = Name_List
Data["Location"] = Location_List
Data["Founded"] = Found_List
Data["Employees"] = Employees_List
Data["About"] = About_List
Data["Skills"] = Skills_List
Data["Link"] = Links_List

#Now Identifying Where and What is Wrong in Data
#print(Data)
```
